# An intuitive approach to Newton’s method

Newton’s method (also known as the Newton-Raphson method) is a very handy way to solve for the roots of a function. Understanding how this technique works is also a great way to understand how and why Gradient Descent makes sense.

<b>Motivation:</b> Say we’d like to find the value of √2. We could set up a polynomial function like the one below:

 <center> <b><i> f(x) = x<sup>2</sup> - 2 </i></b> </center>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Define our function, it's derivative and inverse.
def f(x):
    return np.power(x,2)-2

def fprime(x):
    return 2*x

def finv(y):
    return np.power((y+2),0.5)

def initializePlot(target,start):
    #Set the range of X-axis
    xT,yT = target
    xS,yS = start
    delta = np.abs(xS-xT)/10
    delta = delta if delta > 0.01 else 0.01
    x = np.arange(xT-delta,xS+delta,delta)
   
    #Set up the plot
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

    #Spine placement data centered
    ax.spines['left'].set_position(('data', xT-delta))
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    #Plot our function y=x^2-2
    plt.ylabel('f(x)')
    plt.xlabel('x')
    ax.plot(x,f(x))
    
    #Draw target
    plt.scatter(*zip(*[target]),color='red')
    
    #Draw our starting position
    plt.scatter(*zip(*[start]),color='green')
    return fig
```

When plotted, the function takes on the following form:

```python
#Plot just the function y=x^2-2
x = np.arange(-6,6,0.1)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#Spine placement data centered
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#Plot our function y=x^2-2
plt.ylabel('f(x)')
plt.xlabel('x')
ax.plot(x,f(x))
#Draw start
plt.scatter(*zip(*[(5,f(5))]),color='green')
#Draw target
plt.scatter(*zip(*[(finv(0),0)]),color='red')

plt.show()
```

As we can see, the square root of 2 can be found by solving for the the values of x where the function in intersects the x-axis. In other words, we’re looking for values of x that make the following statement true:

 <center> <b><i> f(x) = 0 </i></b> </center>

This is equivalent to saying:

 <center> <b><i> x<sup>2</sup> - 2 = 0 </i></b> </center>

Let’s show that this is indeed going to give us the value of the square root of 2 by rearranging things a little:

 <center> <b><i> x<sup>2</sup> = 2 </i></b> </center>

and taking the root of both sides

 <center> <b><i> x = √2 </i></b> </center>

If we can find the values of x at which the function crosses the x-axis, we will have found the value of the square root of 2. But the question remains — how do we solve for this value of x? Let’s choose a point (shown in green) on the graph away from the x-intercept: (x=5,f(x)=23)

```python
#Clean up fig from above
plt.close(fig)
```

```python
#target position
yTarget = 0
target = (finv(yTarget),yTarget)
```

```python
#starting position
x1 = 5
y1 = f(x1)
start = (x1,y1)
```

```python
#Draw the tanget to the starting position
#The tangent to the function at any given point
#is simply a line that passes through the chosen point on the function
#and has the same slope as that of the function (at the chosen point).

#So the tangent to our function at the starting point (green point)
#must have the same slope as the function at that point
slopeOfFunctionAtStartingPoint = fprime(x1)
slopeOfTangent = slopeOfFunctionAtStartingPoint

# The slope of a line is given by the rise / run 
# i.e. change in y / change in x
# i.e. f'(x1) = slopeOfTangent =  (y2 - y1) / (x2 - x1) where (x1,y1) and (x2,y2) are points on the line
# Given that slope =  (y2 - y1) / (x2 - x1)
# And (x1,y1) == (xStart,yStart) 
# And (x2,y2) == (xXIntercept,yXIntercept)
# And f(x2) === y2 = 0
# Then we can solve the equation for xXIntercept.
#
# (x2-x1)*f'(x1) = f(x2) - f(x1)
# x2 = {x1*f(x1) -f(x1)}/f'(x1)
# x2 = x1 - f(x1)/f'(x1)
# 
# Once we know the values of x2,y2, and since we already knew the values of x1,y1 we can now draw the line that
# joins the two points.
delta = 1
iteration = 0
while delta > np.power(10.0,-6): #Stop when the difference between y1' and yTarget <= 1x10^-6
    print(f'----Iteration {iteration} (x1,y1) = ({np.round(x1,3)},{np.round(y1)})-------------------------------')
    y1 = f(x1)
    x2 = x1 - f(x1)/fprime(x1)
    y2 = 0
    x1prime = x2
    y1prime = f(x1prime)
    xInterceptPoint = (x2,y2)
    fig = initializePlot(target,(x1,y1))
    plt.scatter(*zip(*[xInterceptPoint]),color='orange',linewidth=0.05)
    plt.plot(*zip(*[xInterceptPoint,(x1,y1)]))
    plt.plot(*zip(*[xInterceptPoint,(x2,f(x2))]),'r--', linewidth=0.5)
    plt.scatter(*zip(*[(x2,f(x2))]),color='black',linewidths=0.05)
    plt.show()
    #plt.savefig(f'./images/{iteration}.png', bbox_inches='tight',dpi=500) #Save images to file
    plt.close(fig)
    delta = np.abs(y1prime - yTarget)
    print(f"|difference| between y1' and yTarget = {delta}\n")
    x1 = x1prime
    y1 = y1prime
    iteration += 1
```
